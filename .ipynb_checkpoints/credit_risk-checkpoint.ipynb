```python
import pandas as pd
```

```python
# Mudra Loan data
mudra = pd.read_csv("C:\\PROJECT LOAN\\mudra_loans.csv")

```

```python
# NSS HouseholdASSets data
nss = pd.read_csv("C:\\PROJECT LOAN\\nss_assets.csv")
```

```python
#cleaning column names
```

```python
mudra.columns = ['sl_no','state_ut','loan_accounts','disbursed_amount_crore']
```

```python
print(nss.columns)
```

```python
nss.columns = [
    'country','state_ut','year','residence_type','household_type',
    'asset_holding_class','asset_type','asset_name','avg_value_assets'
]
```

```python
print(mudra.columns)
```

```python
mudra['disbursed_amount_rs'] = mudra['disbursed_amount_crore'] * 1e7
```

```python
#Merge the datasets
df = pd.merge(
    mudra[['state_ut','loan_accounts','disbursed_amount_rs']],
    nss[['state_ut','household_type','asset_holding_class','asset_name','avg_value_assets']],
    on='state_ut',
    how='inner'
)
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
#Distribution of Loan Accounts & Disbursed Amount
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
df.groupby('state_ut')['loan_accounts'].sum().sort_values().plot(kind='barh')
plt.title("Loan Accounts by State")
plt.show()
```

```python
# Average Debt-to-Asset Ratio by State
```

```python
# Debt-to-Asset Ratio = Disbursed Amount ÷ Avg Household Assets
df['debt_to_asset'] = df['disbursed_amount_rs'] / df['avg_value_assets']
```

```python
# Cash Loan Intensity = Avg Cash Loan Outstanding ÷ Avg Household Assets
df['cash_loan_intensity'] = df.apply(
    lambda x: x['avg_value_assets'] if 'Cash' in str(x['asset_name']) else 0, axis=1
)
```

```python
print(df.shape)          # How many rows and columns?
print(df.columns)        # Which columns exist?
print(df[['state_ut','disbursed_amount_rs','avg_value_assets']].head())
print(df['debt_to_asset'].describe())  # Summary stats
```

```python
import numpy as np

# Avoid division by zero
df['debt_to_asset'] = df['disbursed_amount_rs'] / df['avg_value_assets'].replace(0, np.nan)

# Drop rows where debt_to_asset is infinite or NaN
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna(subset=['debt_to_asset'])
```

```python
state_risk = df.groupby('state_ut', as_index=False)['debt_to_asset'].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
state_risk.sort_values('debt_to_asset').plot(
    x='state_ut', y='debt_to_asset', kind='barh', color='orange', legend=False
)
plt.title("Average Debt-to-Asset Ratio by State")
plt.xlabel("Debt-to-Asset Ratio")
plt.ylabel("State/UT")
plt.show()
```

```python
# Cash Loan Intensity by Household Type

import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
df.groupby('household_type')['cash_loan_intensity'].mean().dropna().plot(kind='bar', color='green')
plt.title("Cash Loan Intensity by Household Type")
plt.xlabel("Household Type")
plt.ylabel("Cash Loan Intensity")
plt.show()
```

```python
# Risk Tier Distribution
#Segment states/households into Low, Medium, High risk tiers.

import pandas as pd

# Create risk tiers
df['risk_tier'] = pd.qcut(df['debt_to_asset'], q=3, labels=['Low','Medium','High'])

# Plot distribution
df['risk_tier'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['green','yellow','red'])
plt.title("Risk Tier Distribution")
plt.ylabel("")
plt.show()

```

```python
#Average Loan Size per State

mudra['avg_loan_size'] = mudra['disbursed_amount_rs'] / mudra['loan_accounts']

plt.figure(figsize=(10,6))
mudra.sort_values('avg_loan_size').plot(x='state_ut', y='avg_loan_size', kind='barh', color='purple', legend=False)
plt.title("Average Loan Size per State")
plt.xlabel("Average Loan Size (Rs)")
plt.ylabel("State/UT")
plt.show()
```

```python
df.to_csv("credit_risk_data.csv", index=False)
```
